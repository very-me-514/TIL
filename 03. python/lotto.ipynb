```python
item = data['data']['list'][0]

# 필요한 정보만 추출하여 딕셔너리로 생성
extracted_info = {
    "draw_no": item.get("ltEpsd"),       # 회차
    "b1": item.get("tm1WnNo"),     # 당첨번호 1
    "b2": item.get("tm2WnNo"),     # 당첨번호 2
    "b3": item.get("tm3WnNo"),     # 당첨번호 3
    "b4": item.get("tm4WnNo"),     # 당첨번호 4
    "b5": item.get("tm5WnNo"),     # 당첨번호 5
    "b6": item.get("tm6WnNo"),     # 당첨번호 6
    "bonus": item.get("bnsWnNo"),     # 보너스 번호
    "money": item.get("rnk1WnAmt")  # 1등 당첨금
}

# 결과 출력
print(extracted_info)
```

```python
keys = {'b1', 'b2', 'b3', 'b4', 'b5', 'b6'}

balls = []
for k in keys:
    balls.append(extracted_info[k])

bonus = extracted_info['bonus']

# 테스트할 때 숫자 바꿔보기
my_number = {1, 7, 9, 17, 27, 38}  # 1등
# my_number = [1, 7, 9, 17, 27, 31]  # 2등
# my_number = [1, 7, 9, 17, 27, 45]  # 3등

# 등수 체커

s1 = set(balls)
s2 = set(my_number)

match_count = len(s1 & s2)
if match_count == 6:
    result = 1
elif match_count == 5 & bonus in my_number:
    result = 3
elif match_count == 5:
    result = 3
elif match_count == 4:
    result = 4
elif match_count == 3:
    result = 5

print(result)


```
