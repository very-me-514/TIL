## 05. 함수



```python
# name: plus_one / in: 숫자 / out: 숫자 +1

def plus_one(num):
    return num + 1

#함수 호출(call)
plus_one(1)
```

```python
# n번 인사하는 함수
# name : hello_ntimes

def hello_ntimes(n):
    # for-매개변수에 들어간 값을 사용하지 않을 떄
    for _ in range(n):
        print('반가워')

hello_ntimes(5)
```

```python
hello_ntimes(10)
```

```python
# 3제곱 해주는 함수
# name : cube, in: 숫자 1개, out : 숫자의 세제곱

def cube(n) :
    return n ** 3

cube(3)
```

```python
# 원기둥의 부피를 구하는 함수
# name : cylinder, in: 반지름, 높이, out: 부피

def cylinder(r, h):
    return (r**2) * h * 3.14

cylinder(2, 5)
```

```python
# 리스트 내부의 숫자들의 평균을 구해주는 함수
# name : get_avg, in : 숫자들로 만들어진 리스트, out : 평균
def get_avg(n):
    return sum(n)/len(n)
    

get_avg([1, 2, 3, 4, 5])
```
