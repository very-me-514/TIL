```python
# 원 (circle)
class Circle : 
    # 생성자 함수 -> 처음 인스턴스가 만들어질 때 자동호출되는 함수.
    def __init__(self, r):
        self.r = r
        print(f'반지름이 {self.r}인 원이 생성되었습니다.')
    
    def get_area(self):
        return 3.14* self.r**2
    
    def __add__(self, other):
        return Circle(self.r + other.r)

c1 = Circle(3)  # 반지름이 3인 원
c2 = Circle(5)  # 반지름이 5인 원

c1.get_area() # 반지름이 3인 원의 넓이
c2.get_area() # r=5의 넓이

print(c1.get_area())
print(c2.get_area())

c3 = c1 + c2   # r=3 + r=5 -> r = 8 원을 새롭게 만들겠다
c3.get_area()
```
