## 04_조건_반복

```python
# 조건_if

n=10

if n > 0: 
    # 블럭 안은 조건이 True일 때만 실행
    print('n은 양수입니다.')
    print('11111')
# 블럭 밖은 조건문과 상관 없음
print('222222')
```

```python
# 조건 - if - else

n = 10

if n > 0:
    print('0보다 크다.')
else:
    print('0보다 작거나 같다.')
```

```python
# 조건 - if - elif - else

n = 0

if n > 0:
    print('양수')
elif n == 0:
    print('0이다')
elif n < 0:
    print('음수')
else:
    print('????')


n = 0

if n > 0:
    print('양수')
elif n == 0:
    print('0이다')
else:
    print('음수')
```

```python
score = 765

# 불합 : 0-399
# 1: 400-599
# 2: 600-749
# 3: 750-899
# 4: 900-1000

if score >= 900:
    grade = 4
elif score >= 750:
    grade = 3
elif score >= 600:
    grade = 2
elif score >= 400:
    grade = 1
else:
    grade = "불합"

print(f'level {grade}')
```

```python
# 우수 학생 판별 (중첩조건문)
status = {
    'score': 80,
    'absent': 5
}

# 1. 점수가 70을 넘는 학생이면 우수학생
# 2. 우수학생 중에 결석일이 5일 미만이면 최우수 학생
# 3. (우수 아닌) 결석일이 5일 미만이면 성실학생
# 4. 나머지는 일반학생

if status['score'] > 70:
    if status['absent'] < 5:
        student_type = '최우수학생'
    else:
        student_type = '우수학생'
elif status['absent'] < 5:
    student_type = '성실학생'
else:
    student_type = '일반학생'

print(student_type)

```

## while 구문

```python
n = 0
while n < 5:
    print('출력', n)
    n += 1

print('끝나고', n)
```

```python
# 

numbers = [1, 2, 3, 4, 5, 6]


total = 0
index = 0 
while index < len(numbers):
    total += numbers[index]
    index += 1

print(total)
```

## for...in
- for 임시변수 in 리스트, 문자열 등등
  리스트나 문자열의 각 요소를 임시변수에 하나씩 대입하면서 블럭을 반복 실행


```python
numbers = [1, 2, 3, 4, 5]
total = 0
for babo in numbers:
    total += babo

print(total)
```

```python
total = 0
for n in range(5):
   total += n
print(total)
```

```python
# 리스트는 idx 수정 가능, 문자열은 idx 수정 불가
s = 'python'

for c in s:
    print(f'!{c}!')
```
