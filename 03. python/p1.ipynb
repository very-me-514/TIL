문제 2. 최대값 찾기
숫자 리스트가 있다.

```python
numbers = [3, 7, 2, 9, 5]

#max(numbers)
#우선 최대값 후보로 첫번째 숫자를 잡는다

maximum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num

print(maximum)
```

문제 6. 재고 확인 시스템

inventory = {
    "apple": 10,
    "banana": 0,
    "orange": 5
}각 과일의 재고가
0이면 → "품절"
1 이상이면 → "판매 가능"
출력하시오.

```python
inventory = {
    "apple": 10,
    "banana": 0,
    "orange": 5
}

total = 0
if inventory['apple'] == 0:
    stock=('품절')
else:
    stock=('판매 가능')
if inventory['banana'] == 0:
    bstock=('품절')
else:
    bstock=('판매 가능')
if inventory['orange'] == 0:
    ostock=('품절')
else:
    ostock=('판매 가능')

print(stock, bstock, ostock)
```

```python
inventory = {
    "apple": 10,
    "banana": 0,
    "orange": 5
}


# 1. 맵핑(Mapping)의 약속
# 딕셔너리는 '키(Key)'와 '값(Value)'이 한 쌍으로 묶여 있습니다.
# 약속: 딕셔너리명[키]라고 쓰면, 그 키에 연결된 값을 즉시 반환한다.
# 마치 사전에서 단어(Key)를 찾으면 그 뜻(Value)이 나오는 것과 같은 원리입니다.

for key in inventory:
    if inventory[key] > 0:
        print(key, '판매가능')
    else:
        print(key, '품절')
```

문제 7. 리스트 안 딕셔너리 평균 구하기
students = [
    {"name": "철수", "score": 80},
    {"name": "영희", "score": 90},
    {"name": "민수", "score": 70}
]
전체 평균 점수 구하기
평균 이상인 학생 이름 출력

```python
students = [
    {"name": "철수", "score": 80},
    {"name": "영희", "score": 90},
    {"name": "민수", "score": 70}
]

total = 0
for student in students:
    total +=student['score']

# 전체 평균 점수
avg = total / len(students)
print('평균점수', avg)

# 평균 이상인 학생 출력
for student in students:
    if student['score'] >= avg:
        print(student['name'])
```

문제 8. 등급 매기기

점수 리스트:

scores = [95, 82, 67, 74, 88]

각 점수에 대해

90 이상 → A
80 이상 → B
70 이상 → C
그 외 → D
등급 리스트를 새로 만들어 출력하시오.

```python
scores = [95, 82, 67, 74, 88]

grades=[] 
index = 0 
for score in scores:
    if score >= 90:
        grades.append('A')
    elif score >= 80:
        grades.append('B')
    elif score >= 70:
        grades.append('C')
    else:
        grades.append('D')

print(grades)
```

```python
# 문제 10. 비밀번호 검사기
password = "abc123"조건:
길이가 8자 이상이고
숫자가 포함되어 있으면 "사용 가능"
아니면 "사용 불가"
출력하시오.
(힌트: len(), in  사용. for 문으로 한글자씩 체크) (편집됨)
```

```python
password = "abc123"

# 틀렸다고 가정
is_ok = '사용불가'

# 8글자 이상
if len(password) >= 8 :
    # 한글자씩 돌면서 숫자 체크
    for char in password:
        if char.isdigit():
            is_ok = '사용가능'

print(is_ok)
    
    
```
