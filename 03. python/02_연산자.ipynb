### Short-Circuit Evaluation (단축평가)
- 앞에서 값이 확실시 되면, 그 뒤는 확인하지 않는다.
- 조건 판별시 뒷부분을 확인하지 않아도 됨

```python
a = True and True and True and False and True and False
#     ?   ->  ?   ->   ?   ->   !  [종료, 뒤에 볼 필요 없음]
b = False or False or True or False
#     ?   ->  ?   ->   !

print(a, b)
```

## 복합 연산자
- 연산 + 대입을 동시에 한다

|연산자	|내용|
|---------|-----|
|a += b|	a = a + b|
|a -= b|	a = a - b|
|a *= b|	a = a * b|
|a /= b|	a = a / b|
|a //= b|	a = a // b|
|a %= b|	a = a % b|
|a**= b	|a = a ** b|


```python
x = 1
x = x + 1
x += 1
x
```

```python
s = 'py'
s += 'thon'

s
```

```python
# bmi 계산기
# bmi = 몸무게(kg) / 키(m) 제곱 2
# weight 에 몸무게, height 에 키를 저장하고 bmi 를 계산하기

weight = 70
height = 1.75
bmi = weight / (height ** 2)
print(bmi)
```

```python
# 원 넓이와 둘레 계산 (pi 는 3.14 로 계산)
# 원 넓이 = pi x r^2
# 원 둘레 = 2 x pi x r

from math import pi

r = 3
pi = 3.14

a = pi * (r ** 2)
c = 2 * pi * r

print(a, c)
```

```python
# 초를 시간, 분, 초 로 바꾸기
# 계산을 통해 h, m, s 에 시간, 분, 초로 전환

seconds = 3675  # 1시간 1분 15초

time = seconds // 3600
minute = seconds % 3600 // 60
seconds = seconds % 3600 % 60
print(time, minute, seconds)
```
