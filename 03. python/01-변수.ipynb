### f-string (f '', f(''''''))
string 안에 변수를 녹여서 써야할 때 사용

```python
name = '김광주'
Age = 30

# f-string (f '')
greet = name + '님 환영합니다.'
greeting = f'{name}님 환영합니다. \n{name}님의 나이는 {Age}살 입니다.'  # 김광주 님 환영합니다. 30살입니다.

print(greeting) 
```

```python
# string 안에서 따옴표 사용하기
title = "lily\'s bag"
story = '그는 말했다. \"안녕?\" 이라고.'

print(title)
print(story)
```
